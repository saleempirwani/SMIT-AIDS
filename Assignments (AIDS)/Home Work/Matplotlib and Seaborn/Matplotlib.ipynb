```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x **2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.plot(y, x, 'r')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'r')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(y, x, 'r')
axes2.set_xlabel('X label2')
axes2.set_ylabel('Y label2')
axes2.set_title('Title2')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First plot')

axes[1].plot(x, y)
axes[1].set_title('Second plot')


plt.tight_layout()
```

### Figure and DPI (dot per inches)

```python
fig = plt.figure(figsize=(8, 2)) 

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2)) 

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig.savefig("plot.png", dpi=200)
```

```python
fig = plt.figure() 

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x**2, label='Square')
axes.plot(x, x**3, label='Cube')

plt.legend(bbox_to_anchor=(1, 1), title="Legends")
```

```python
fig = plt.figure() 

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y, color="purple", lw=3, ls="-", marker="o", markersize=10, 
          markerfacecolor="yellow", markeredgecolor="green", markeredgewidth=3)
```

```python
fig = plt.figure() 

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)

axes.set_xlim([0, 1])
axes.set_ylim([0, 2])
```

```python
x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

#### Exercises # 1

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('X-Y Title')
ax.plot(x, y)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax.plot()
ax2.plot()
```

```python
ax.plot(x, y)
ax2.plot(x, y)

fig
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax.plot()
ax2.plot()
```

```python
ax.plot(x, y, z)
ax.set_xlim([0, 100])
ax.set_ylim([0, 10000])
ax.set_xlabel('x')
ax.set_ylabel('z')


ax2.plot(x, y, z)
ax2.set_xlim([20, 22])
ax2.set_ylim([300, 500])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')

fig
```
