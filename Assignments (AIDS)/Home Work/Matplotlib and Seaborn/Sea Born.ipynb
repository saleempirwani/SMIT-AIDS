## Distribution Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
sns.histplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')  
# kind='hex' OR 'reg'
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
# hue='sex'(categorical column, means col has category)
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
# kde == Kernel Density Estimation
```

## Categorical Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=sum)

# estimator = sum, len, np.mean (some aggregate function)
```

```python
sns.countplot(x='sex', data=tips)
# same as pd.DataFrame.count()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')

# recommends for CEO type person.
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', 
              data=tips, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')

# sns.factorplot == sns.catplot
```

### Matrix Plots

```python
flights.head()    
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(data=tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```python
sns.heatmap(fp, cmap='coolwarm', linewidths=1, linecolor='white')
```

```python
sns.clustermap(data=fp, cmap='coolwarm', standard_scale=1)
```

## Grid Plots

```python
 iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, row='time', col='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

## Regression Plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips, 
           hue='sex', markers=['o', 'v'], 
           scatter_kws={'s': 100}, palette='seismic')

# scatter_kws => for increase of markers.
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, 
           col='day', hue='sex', aspect=1, height=8)
```
