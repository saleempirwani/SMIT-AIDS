# Numpy (Cont...)

```python
import numpy as np
```

```python
arr1 = np.arange(16).reshape(8,2)
arr1
```

```python
arr2 = np.arange(16,56).reshape(8,5)
arr2
```

```python
np.concatenate((arr1,arr2),axis=1)   #similar to hstack
```

```python
arr1 = np.arange(32).reshape(8,4)
arr1
```

```python
arr2 = np.arange(16).reshape(4,4)
arr2
```

```python
np.concatenate((arr1,arr2),axis=0)   #similar to vstack
```

### Split

```python
x = np.arange(9.0)
x
```

```python
np.split(x, 3)    # horizontally
```

```python
a = np.arange(12).reshape(3, 4)
a
```

```python
# horizon split the 2-d array into 4 array columns
np.hsplit(a, [1,3])
```

```python
# split at columns 1 and 3
np.vsplit(a, 3)
```

### Manipulating and Saving the Image

```python
import numpy as np
from PIL import Image

myImg = Image.open('car.jpg')
myImg = myImg.convert('L')
myImg = np.array(myImg)
#myImg = np.array(Image.open('kolala.jpeg').convert())

print(type(myImg))

gr_im = Image.fromarray(myImg).save('gr_car.png')
gr_im
```
